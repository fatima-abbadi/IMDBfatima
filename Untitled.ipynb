```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv('imdb_top_1000.csv')
print(df)
```

```python


# Handle missing values
df = df.dropna(axis=0)
print(df)

```

```python
# Count the occurrences of each genre
genre_counts = df['Genre'].value_counts()

# Print the most popular genre
most_popular_genre = genre_counts.idxmax()
print("The most popular genre:", most_popular_genre)
```

```python
# Count the occurrences of each genre
genre_counts = df['Genre'].value_counts()

# Create a new dataframe with the genre counts
genre_counts_df = pd.DataFrame(genre_counts).reset_index()
genre_counts_df.columns = ['Genre', 'Count']

# Print the dataframe
print(genre_counts_df)
```

```python


# Create a figure with custom size
plt.figure(figsize=(100, 20))

# Create a bar chart
plt.bar(genre_counts_df['Genre'], genre_counts_df['Count'])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of  Shows/Fatima Abbadi')
plt.xticks(rotation=90)

# Show the chart
plt.show()

```

```python
# Count the occurrences of each director
director_counts = df['Director'].value_counts()

# Create a new dataframe with the director counts
director_counts_df = pd.DataFrame(director_counts).reset_index()
director_counts_df.columns = ['Director', 'Count']

# Sort the dataframe by count in descending order
director_counts_df = director_counts_df.sort_values(by='Count', ascending=False)

# Print the dataframe
print(director_counts_df)

```

```python

# Create a bar chart
plt.bar(director_counts_df['Director'], director_counts_df['Count'])
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Top Rated Movies/TV Shows by Director')
plt.xticks(rotation=90)

# Show the chart
plt.show()
```
